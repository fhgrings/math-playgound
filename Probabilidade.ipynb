```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Data for plotting
mu, sigma, size = 50, 20, 1000 # mean and standard deviation
s = np.random.normal(mu, sigma, size)
fig, ax = plt.subplots()
ax.plot(s)
ax.tick_params()
s_sorted = s
s_sorted.sort()
plt.plot(s)
s_avg = sum(s)/len(s)
plt.plot([s_avg]*size)

plt.show()
median = [((x-s_avg)**2)/2 for x in s]
plt.plot(median)
desvio = [math.sqrt(i) for i in median]
plt.plot(desvio)

plt.show()
```

```python
summed = []
for item in s:
    if not len(summed):
        summed.append(item)
        continue
    summed.append(summed[-1] + item)
    
plt.plot(summed)
plt.show()
```

```python
bins = list(range(int(min(s)), int(max(s)), len(s)//(mu//2)))
# plt.hist(s, bins=bins)
count, bins, ignored = plt.hist(s, 30, density=True)

plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2/(2*sigma**2)),
         linewidth=2, color='r')

plt.axvline(x = mu, color = 'b')

median = sum([((x-s_avg)**2)/2 for x in s])/size
desvio = math.sqrt(median)

plt.axvline(x = mu-desvio, color = 'g')
plt.axvline(x = mu+desvio, color = 'g')

plt.axvline(x = mu-2*desvio, color = 'y')
plt.axvline(x = mu+2*desvio, color = 'y')

plt.axvline(x = mu-3*desvio, color = 'gray')
plt.axvline(x = mu+3*desvio, color = 'gray')


plt.show()
```
