```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
mu, sigma = 2000, 20 # mean and standard deviation
s = np.random.normal(mu, sigma, mu)
fig, ax = plt.subplots()
ax.plot(s)
ax.tick_params()
s_sorted = s
s_sorted.sort()
plt.plot(s)
s_avg = sum(s)/len(s)
plt.plot([s_avg]*mu)


plt.show()
```

```python
summed = []
for item in s:
    if not len(summed):
        summed.append(item)
        continue
    summed.append(summed[-1] + item)
    
plt.plot(summed)
plt.show()
```

```python
plt.hist(s, bins=list(range(int(min(s)), int(max(s)), len(s)//(mu//2))))
plt.show()
```
